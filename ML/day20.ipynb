```python
# Now if we want to compare two columns 
def logic(row):
    profit_percent= (row['earning'] - row['cost']) / row['duration']    
    if profit_percent >=3:
        return row['status'] =='blockbuster'
    elif profit_percent >=2 :
        return row['status'] == 'superhit'
    elif profit_percent >=1 :
        return row['status'] == 'hit'
    elif profit_percent >=0 :
        return row['status'] == 'average'
    else:
        return row['status'] == 'flop'
    
wf['validation'] = wf.apply (logic,axis=1)
wf.head()
```
